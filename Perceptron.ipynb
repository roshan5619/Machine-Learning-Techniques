```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
import pandas as pd
wine_data=pd.read_csv("C:/Users/rosha/Documents/B.Tech/Semester(6)/ML_Techniques_CSE-3008/LAB/Lab-2/winequality-red.csv",sep=";")
wine_data.head()
```

```python
x=wine_data.iloc[:,:-1]
y=wine_data.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import Perceptron
perceptron = Perceptron()
```

```python
perceptron.fit(X_train , y_train)
```

```python
y_pred = perceptron.predict(X_test)
```

```python
print(f"Perceptron accuracy:",accuracy_score(y_test ,y_pred))
```

```python
cm = confusion_matrix(y_test , y_pred)
```

```python
sns.heatmap(cm ,annot =True ,cmap ='Blues' ,fmt ="d")
plt.title("Confusion Matrix")
plt.xlabel("Predicted label")
plt.ylabel("True Label")
plt.show()
```

```python
plt.hist(y_pred , bins=range(3,9),alpha =0.5)
plt.legend()
plt.xlabel("Predicted Wine Quality")
plt.ylabel("Frequency")
plt.title("Histogram of predicted wine quality values")
plt.show()
```
